```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
lp = np.linspace(-3,3,50)
gp = np.linspace(-3,3,20);
x,y = np.meshgrid(gp,(gp+3)/2)
z = x*y
z2 = np.exp(lp*lp/2)*0.05
```

```python
plt.figure(figsize=(20,20))

plt.contour(x,y,z,[1,2,3])
plt.ylim(0,3)
plt.plot(lp,z2)
```
